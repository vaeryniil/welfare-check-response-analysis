```python
#filtering each of the data sets to only include welfare check calls (requests for a welfare check)
WELFARE2015 = SPR2015[SPR2015['Initial Call Type'] == 'CHKWLF']
WELFARE2016 = SPR2016[SPR2016['Initial Call Type'] == 'CHKWLF']
WELFARE2017 = SPR2017[SPR2017['Initial Call Type'] == 'CHKWLF']
WELFARE2018 = SPR2018[SPR2018['Initial Call Type'] == 'CHKWLF']
WELFARE2019 = SPR2019[SPR2019['Initial Call Type'] == 'CHKWLF']
WELFARE2020 = SPR2020[SPR2020['Initial Call Type'] == 'CHKWLF']
WELFARE2021 = SPR2021[SPR2021['Initial Call Type'] == 'CHKWLF']
WELFARE2022 = SPR2022[SPR2022['Initial Call Type'] == 'CHKWLF']
WELFARE2023 = SPR2023[SPR2023['Initial Call Type'] == 'CHKWLF']
WELFARE2024 = SPR2024[SPR2024['Initial Call Type'] == 'CHKWLF']
WELFARE2025 = SPR2025[SPR2025['Initial Call Type'] == 'CHKWLF']
```

```python
#combining into a single SPR welfare data frame [still with no primary units or disposition]
frames = [WELFARE2015, WELFARE2016, WELFARE2017, WELFARE2018, WELFARE2019, WELFARE2020,
          WELFARE2021, WELFARE2022, WELFARE2023, WELFARE2024, WELFARE2025]

WELFARE = pd.concat(frames)
WELFARE
```

```python
#combining into a single SPR unit data frame
frames = [UNIT2015, UNIT2016, UNIT2017, UNIT2018, UNIT2019, UNIT2020,
          UNIT2021, UNIT2022, UNIT2023, UNIT2024, UNIT2025]
units_all_years = pd.concat(frames)
units_all_years.drop(['Unnamed: 1','call time'],axis=1,inplace=True)
units_all_years
```

```python
#try left joining - do the times match? #the call times match. So can delete from original df
intermediate = pd.merge(WELFARE,units_all_years, left_on='Incident Number', right_on='inci id',how='left')
intermediate.drop('inci id',axis=1,inplace=True)
intermediate
```

```python
#combining into a single SPR closed_as data frame
frames = [CLOSE2015, CLOSE2016, CLOSE2017, CLOSE2018, CLOSE2019, CLOSE2020,
          CLOSE2021, CLOSE2022, CLOSE2023, CLOSE2024, CLOSE2025]
closed_all_years = pd.concat(frames)
closed_all_years
```

```python
SPR_welfare = pd.merge(intermediate,closed_all_years, on='Incident Number',how='left')
SPR_welfare
```

```python
#Make new columns for month and year, for later grouping
SPR_welfare['month'] = SPR_welfare['Call Creation Time'].dt.month
SPR_welfare['year'] = SPR_welfare['Call Creation Time'].dt.year
SPR_welfare
```

```python
#Strip whitespace
SPR_welfare['Primary Responding Unit'] = SPR_welfare['Primary Responding Unit'].str.strip()
SPR_welfare['prime unit'] = SPR_welfare['prime unit'].str.strip()
SPR_welfare['units dispatched in order'] = SPR_welfare['units dispatched in order'].str.strip()
```

```python
#Make an institution column based on primary responding unit
#checking about null values 1st
nullprime = SPR_welfare[SPR_welfare['Primary Responding Unit'].isnull()]
nullprime
```

```python
#nullprime[nullprime['units dispatched in order'].isna()] #3434 rows
#nullprime.shape #3966
nullprime[nullprime['prime unit'].isna()] #3966
#units_sent = nullprime[(nullprime['units dispatched in order'].notna()) & (units_sent['units dispatched in order'] == '3J81')] #532
units_sent = nullprime[nullprime['units dispatched in order'].notna()]
units_sent
```

```python
units_sent['Close Code'].unique()
#so, apparantly, some of these calls with no primary unit did actually end in welfare checks
```

```python
#getting the count of units dispatched

#first, count no_of_units = SPR_welfare['units dispatched in order 
unit_count = SPR_welfare['units dispatched in order'].str.count(",") + 1


#second, make new column
SPR_welfare['units dispatched'] = unit_count

#third, convert NA values to ''
SPR_welfare['units dispatched'] = SPR_welfare['units dispatched'].fillna(0)


SPR_welfare
```

```python
unit_count2 = nullprime['units dispatched in order'].str.count(",") + 1
nullprime['units dispatched'] = unit_count2
nullprime['units dispatched'] = nullprime['units dispatched'].fillna(0)
nullprime
```

```python
#dropping rows where the primeunit is a null value and more than 0 units were dispatched
SPR_welfare = SPR_welfare[(SPR_welfare['Primary Responding Unit'].notna()) | (SPR_welfare['units dispatched'] == 0.)]
```

```python
#Adding a column saying whether cahoots, spd, or neither answered the call

SPR_welfare["Institution"] = np.where(    
    SPR_welfare["Primary Responding Unit"].str.contains("J|CAH", na=False),    
    "CAHOOTS",    
    np.where(        
        SPR_welfare["Primary Responding Unit"].isna(),        
        "Neither",        
        "SPD"    
    )
)
```
